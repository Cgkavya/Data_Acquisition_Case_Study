## Q6. Write code to get the list of files in your Downloads folder and save it to a CSV file name "download_list.csv"

```python

import csv
# Get the path to the Downloads folder
downloads_folder = os.path.expanduser("~/Downloads")

# Create the full path for the CSV file
csv_file_path = os.path.join(os.getcwd(), "download_list.csv")

# Get the list of files in the Downloads folder
files_in_downloads = os.listdir(downloads_folder)

if files_in_downloads:
   try:
       with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(["File Name"]) # Write header
            for file_name in files_in_downloads:
                writer.writerow([file_name])
       print(f"File list saved to '{csv_file_path}' successfully.")
   except IOError as e:
        print(f"Error writing to CSV file: {e}")
else:
    print("No files found in the Downloads folder or an error occurred.")
```

## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
import matplotlib.pyplot as plt

data = data.drop('instant', axis = 1)

# Identify the numerical columns
num_df = data.select_dtypes(include = "number")

num_df.boxplot()
plt.xticks(rotation = 45)
plt.show()

# IQR approach for finding outliers
Q1 = data['registered'].quantile(0.25)
Q3 = data['registered'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(IQR)
print(lower_bound, upper_bound)

outliers = data[(data['registered'] < lower_bound) | (data['registered'] > upper_bound)]

print(outliers[['registered']])
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
# Mapping of season numbers to names
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}

# Group by season and sum the total rental count ('cnt')
rentals_by_season = data.groupby('season')['cnt'].sum()

# Find the season with the maximum rentals
most_rented_season_num = rentals_by_season.idxmax()
most_rented_season_name = season_map[most_rented_season_num]

print(most_rented_season_name )
```

## Q.17 What is the peak hour in which bike rents the most?

```python
# Group by hour and sum the total rental count ('cnt')
rentals_by_hour = data.groupby('hr')['cnt'].sum()

# Find the hour with the maximum rentals
peak_hour = rentals_by_hour.idxmax()

print(peak_hour)
```

## Q.18 In which day of a week bikes rents out most?

```python
# Mapping of weekday numbers to names (0: Monday, 6: Sunday)
weekday_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}

# Group by weekday and sum the total rental count ('cnt')
rentals_by_weekday = data.groupby('weekday')['cnt'].sum()

# Find the weekday with the maximum rentals
most_rented_weekday_num = rentals_by_weekday.idxmax()
most_rented_weekday_name = weekday_map[most_rented_weekday_num]

print(most_rented_weekday_name)
```

## Q.19 In which hour Casual users rents bikes the most?

```python
# Group by hour and sum the casual user count ('casual')
casual_rentals_by_hour = data.groupby('hr')['casual'].sum()

# Find the hour with the maximum casual rentals
peak_hour_casual = casual_rentals_by_hour.idxmax()

print(peak_hour_casual)
```

## Q.20 What is the maximum temperature observed in each of the seasons?

```python
# Group by season and find the maximum temperature ('temp') in each season
max_temp_by_season = data.groupby('season')['temp'].max()

print(max_temp_by_season)
```
